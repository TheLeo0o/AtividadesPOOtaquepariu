```python
from abc import ABC, abstractmethod

class MotorVehicle(ABC):
    @abstractmethod
    def build(self):
        pass

class Motorcycle(MotorVehicle):
    def build(self):
        print("Contruindo uma motocicleta ðŸï¸")
    
class Car(MotorVehicle):
    def build(self):
        print("Contruindo um carro ðŸš—")

class MotorVehicleFactory(ABC):
    @abstractmethod
    def create_motor_vehicle(self) -> MotorVehicle:
        pass

class MotorcycleFactory(MotorVehicleFactory):
    def create_motor_vehicle(self) -> MotorVehicle:
        return Motorcycle()

class CarFactory(MotorVehicleFactory):
    def create_motor_vehicle(self) -> MotorVehicle:
        return Car()

if __name__ == "__main__":
    motorcycle_factory = MotorcycleFactory()
    motorcycle = motorcycle_factory.create_motor_vehicle()
    motorcycle.build()

    car_factory = CarFactory()
    car = car_factory.create_motor_vehicle()
    car.build()
```

```python
from abc import ABC, abstractmethod

class Product(ABC):
    def __init__(self, color):
        self.color = color
    
    def get_color(self):
        return self.color

    @abstractmethod
    def __str__(self):
        pass

class Armchair(Product):
    def __init__(self,color, width, height, depth):
        super().__init__(color)
        self.width = width
        self.height = height
        self.depth = depth

class CoffeeTable(Product):
     def __init__(self,color,width, height, length):
        super().__init__(color)
        self.width = width
        self.height = height    
        self.length = length

class Sofa(Product):
    def __init__(self,color, width, height, depth):
        super().__init__(color)
        self.width = width
        self.height = height
        self.depth = depth

class ModernArmchair(Armchair):
    def __init__(self):
        super().__init__("Black", 80,90,85)

    def __str__(self):
        return f"Modern Armchair: Color: {self.color}, Size: {self.width}x{self.height}x{self.depth}"    

class ModernCoffeeTable(CoffeeTable):
    def __init__(self):
        super().__init__("White", 120, 45, 60)
    
    def __str__(self):
        return f"Modern Coffee Table: Color: {self.color}, Size: {self.width}x{self.height}x{self.length}"

class ModernSofa(Sofa):
    def __init__(self):
        super().__init__("Gray", 200, 100, 95)
    
    def __str__(self):
        return f"Modern Sofa: Color: {self.color}, Size: {self.width}x{self.height}x{self.depth}"       

class FurnitureFactory(ABC):
    @abstractmethod
    def make_armchair(self) -> Armchair:
        pass

    @abstractmethod
    def make_coffee_table(self) -> CoffeeTable:
        pass

    @abstractmethod
    def make_sofa(self) -> Sofa:
        pass

class RetroFurnitureFactory(FurnitureFactory):
    def make_armchair(self):
        return ModernArmchair()
    
    def make_coffee_table(self):
        return ModernCoffeeTable()
    
    def make_sofa(self):
        return ModernSofa()
    
if __name__ == "__main__":
    factory = RetroFurnitureFactory()

    armchair = factory.make_armchair()
    coffee_table = factory.make_coffee_table()
    sofa = factory.make_sofa()

    print(armchair)
    print(coffee_table)
    print(sofa)
```
